```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
import json
import seaborn as sns
import re
from IPython.core.display import HTML
```

# Dauphin Island Restaurants

```python
URL = 'https://www.townofdauphinisland.org/restaurants'
response = requests.get(URL)
```

```python
soup = BS(response.text)
```

```python
restaurants = soup.findAll('span')
restaurants
```

```python
restaurants_list = []

for x in soup.select('span'):
    restaurants_list.append(x.text)
```

```python
restaurants_list = restaurants_list[4:]
restaurants_list
```

```python
restaurants_list = restaurants_list[:-9]
```

```python
dauphin_restaurants = pd.DataFrame({'restaurant' :restaurants_list})
dauphin_restaurants = dauphin_restaurants[dauphin_restaurants['restaurant'].str.contains('Read More')==False].reset_index(drop=True)
dauphin_restaurants
```

```python
sites = soup.findAll('a', attrs={'aria-disabled':'false'})
sites
```

```python
website_list = []

for x in sites:
    website_list.append(x.get('href'))
```

```python
website_list = website_list[:-4]
```

```python
dauphin_restaurants['website'] = website_list
dauphin_restaurants
```

```python
def get_description(url):
    response = requests.get(url)
    soup = BS(response.text)
    description = soup.findAll('p')
    return description[7].text
```

```python
get_description(dauphin_restaurants.loc[0, 'website'])
```

```python
dauphin_restaurants['description'] = dauphin_restaurants.website.apply(get_description)
```
