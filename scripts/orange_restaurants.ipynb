```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
import json
import seaborn as sns
import re
from IPython.core.display import HTML
```

```python
URL = 'https://familydestinationsguide.com/best-restaurants-in-orange-beach-al/'
response = requests.get(URL)
```

```python
soup = BS(response.text)
soup
```

```python
restaurants_list = []

for x in soup.select('h2'):
    restaurants_list.append(x.text)
```

```python
restaurants_list = restaurants_list[:-4]
restaurants_list
```

```python
orange_restaurants = pd.DataFrame({'restaurants':restaurants_list})
```

```python
orange_restaurants[['code', 'restaurant']] = orange_restaurants['restaurants'].str.split(' ', 1, expand=True)
orange_restaurants = orange_restaurants.drop(columns = {'restaurants', 'code'})
orange_restaurants
```
