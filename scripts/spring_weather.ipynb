# All Spring Weather (minus 2023)

```python
spring_weather = pd.merge(spring_2013, spring_2014,
        on=('name', 'datetime', 'tempmax', 'tempmin', 'temp', 'humidity', 'precipprob', 'preciptype', 'windspeed', 'conditions', 'location'),
        how='outer')
spring_weather = pd.merge(spring_weather, spring_2015,
        on=('name', 'datetime', 'tempmax', 'tempmin', 'temp', 'humidity', 'precipprob', 'preciptype', 'windspeed', 'conditions', 'location'),
        how='outer')
spring_weather = pd.merge(spring_weather, spring_2016,
        on=('name', 'datetime', 'tempmax', 'tempmin', 'temp', 'humidity', 'precipprob', 'preciptype', 'windspeed', 'conditions', 'location'),
        how='outer')
spring_weather = pd.merge(spring_weather, spring_2017,
        on=('name', 'datetime', 'tempmax', 'tempmin', 'temp', 'humidity', 'precipprob', 'preciptype', 'windspeed', 'conditions', 'location'),
        how='outer')
spring_weather = pd.merge(spring_weather, spring_2018,
        on=('name', 'datetime', 'tempmax', 'tempmin', 'temp', 'humidity', 'precipprob', 'preciptype', 'windspeed', 'conditions', 'location'),
        how='outer')
spring_weather = pd.merge(spring_weather, spring_2019,
        on=('name', 'datetime', 'tempmax', 'tempmin', 'temp', 'humidity', 'precipprob', 'preciptype', 'windspeed', 'conditions', 'location'),
        how='outer')
spring_weather = pd.merge(spring_weather, spring_2020,
        on=('name', 'datetime', 'tempmax', 'tempmin', 'temp', 'humidity', 'precipprob', 'preciptype', 'windspeed', 'conditions', 'location'),
        how='outer')
spring_weather = pd.merge(spring_weather, spring_2021,
        on=('name', 'datetime', 'tempmax', 'tempmin', 'temp', 'humidity', 'precipprob', 'preciptype', 'windspeed', 'conditions', 'location'),
        how='outer')
spring_weather = pd.merge(spring_weather, spring_2022,
        on=('name', 'datetime', 'tempmax', 'tempmin', 'temp', 'humidity', 'precipprob', 'preciptype', 'windspeed', 'conditions', 'location'),
        how='outer')
```

```python
spring_weather['month_day'] = spring_weather['datetime'].str.slice(-5)
```

```python
spring_weather
```

```python
spring_weather_avgs = spring_weather.groupby(['location', 'month_day'])['temp', 'tempmax', 'tempmin', 'humidity', 'windspeed', 'precipprob'].mean().reset_index()
```

```python
spring_weather_avgs['precipprob'] = spring_weather_avgs['precipprob']/10
spring_weather_avgs = spring_weather_avgs.rename(columns = {'precipprob':'days_with_rain_10_years'})
```

```python
clear_weather = spring_weather.loc[spring_weather['conditions']=='Clear'].groupby(['location', 'month_day'])['name'].count().to_frame().reset_index().rename(columns={'name':'days_clear_10_years'})
```

```python
spring_weather_avgs = pd.merge(spring_weather_avgs, clear_weather,
        on=('location', 'month_day'),
        how='inner')
```
